```python
# prompt: colocar un arreglo de (5,2) donde en la primera columna coloque la primera es el valor unitario y la segunda es la cantidad

import numpy as np

# Crear un arreglo de (5, 2)
data = np.array([
    [10, 2],  # Valor unitario: 10, Cantidad: 2
    [5, 5],   # Valor unitario: 5, Cantidad: 5
    [20, 1],  # Valor unitario: 20, Cantidad: 1
    [15, 3],  # Valor unitario: 15, Cantidad: 3
    [8, 4]    # Valor unitario: 8, Cantidad: 4
])

data
```

```python
# prompt: Multiplicar las dos columnas del arreglo data y agregarlo para quedar en la forma (5,3)

# Multiplicar las dos columnas del arreglo data
multiplicacion = data[:, 0] * data[:, 1]

# Crear un nuevo arreglo con la forma (5, 3)
nuevo_data = np.zeros((5, 3))

# Copiar los datos originales a las primeras dos columnas del nuevo arreglo
nuevo_data[:, :2] = data

# Agregar el resultado de la multiplicación a la tercera columna
nuevo_data[:, 2] = multiplicacion

nuevo_data
```

```python
# prompt: calcular el promedio de la columna 0 de data

# Assuming 'data' is the NumPy array defined in the previous code.
# If 'data' is a pandas DataFrame, use data.iloc[:, 0].mean() instead.

promedio_columna_0 = np.mean(data[:, 0])
print(f"El promedio de la columna 0 es: {promedio_columna_0}")
```

```python
# prompt: Creame un dataframe con el nombre, el cargo, el salario y la antiguedad en años, poner 20 filas con datos sintéticos

import pandas as pd
import numpy as np

# Crear datos sintéticos
np.random.seed(42)  # Para reproducibilidad (datos aleatorios, punto de arranque)

num_rows = 20
nombres = ['Nombre_' + str(i) for i in range(1, num_rows + 1)]
cargos = np.random.choice(['Analista', 'Ingeniero', 'Gerente', 'Desarrollador'], size=num_rows)
salarios = np.random.randint(30000, 100000, size=num_rows)
antiguedades = np.random.randint(1, 10, size=num_rows)

# Crear el DataFrame
df = pd.DataFrame({
    'Nombre': nombres,
    'Cargo': cargos,
    'Salario': salarios,
    'Antiguedad': antiguedades
})

df
```
